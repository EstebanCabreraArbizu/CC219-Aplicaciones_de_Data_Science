## 1.- Ejercicios de Nomalización:

* Leemos el corpus y pasamos los documentos (Título + Texto) a una lista

```python
docs_file = 'corpus_mundo_today.csv'
docs_list = list()
file_txt = open(docs_file, encoding="utf8").read()
for line in file_txt.split('\n'):
    line = line.split('||')
    docs_list.append(line[1] + ' ' + line[2])
docs_list = docs_list[1:] # Elimino la cabecera del fichero
```

```python
docs_list
```

#### 1. **Crear una función que devuelva los documentos *Tokenizados* (una lista de listas) y con los tokens (palabras) en minúsculas.**

* **input**: lista de documentos (lista de Strings).
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
!python3 -m spacy download es
```

```python
import spacy
nlp = spacy.load('es_core_news_sm')
```

```python
def tokenization(docs_list):
    tokens = []
    for doc in nlp.pipe(docs_list):
        tokens.extend([w.text.lower() for w in doc])
    return tokens
```

```python
lista = tokenization(docs_list)
print(lista)
```

#### 2. **Crear una función que elimine los tokens que sean signos de puntuación y *Stop-Words*.**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation

def remove_words(tokens_list):
    # Unir la lista de tokens en un solo string
    doc_text = ' '.join(tokens_list).lower()

    # Procesar el texto con spaCy
    doc = nlp(doc_text)

    # Obtener stopwords de spaCy
    spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS

    # Eliminar signos de puntuación y stopwords
    tokens_without_punctuation = [token.text for token in doc if token.text not in punctuation]
    filtered_tokens = [token for token in tokens_without_punctuation if token.lower() not in spacy_stopwords]

    return filtered_tokens
```

```python
lista = remove_words(docs_list)
print(lista)
```

#### 3. **Crear una función que transforme cada token a su lema (*Lematización*)**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
def lemmatization(doc_tokens):
    lemmas = []
    for token in doc_tokens:
        doc = nlp(token)
        lemmas.append(doc[0].lemma_)  # Obtiene el lema del primer token en el documento
    return lemmas
```

```python
lista = lemmatization(docs_list)
print(lista)
```

#### 4. **Crear una función que elimine todos los tokens que no sean *Nombres* (NOUN, PROPN), *Verbos*, *Advervios* o *Adjetivos*.**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
def filter_words(tokens_list):
    allowed_pos = {'NOUN', 'PROPN', 'VERB', 'ADV', 'ADJ'}  # Etiquetas POS permitidas

    # Unir la lista de tokens en un solo string
    doc_text = ' '.join(tokens_list)

    # Procesar el texto con SpaCy
    doc = nlp(doc_text)

    # Filtrar tokens por etiquetas POS permitidas
    filtered_tokens = [token.text for token in doc if token.pos_ in allowed_pos]

    return filtered_tokens
```

```python
lista = filter_words(docs_list)
print(lista)
```

#### 5. **Función que dada una lista de documentos, devuelva los documentos normalizados. Este ejercicio ya esta hecho y simplemente tiene que funcionar llamando a las 4 funciones anteriores.**

* **input**: lista de documentos (lista de Strings).
* **output**: lista de listas, en la que cada lista contiene los tokens del documento normalizados.

```python
def normalization(docs_list):
    corpus = tokenization(docs_list)
    corpus = remove_words(corpus)
    corpus = lemmatization(corpus)
    corpus = filter_words(corpus)
    return corpus

corpus = normalization(docs_list)
print(corpus[0])
```

```python
print(corpus)
```

#### 6. **Crear una función que dada una lista de documentos (*corpus*) tokenizados, elimine del corpus aquellos tokens que aparecen menos de 'N' veces (N=10) en el corpus**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento normalizados.
* **input**: 'N' -> Parámetro que nos indica el número mínimo de apariciones de la palabra en el corpus.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento normalizados.

```python
from collections import Counter

from collections import Counter

def drop_less_frecuency_words(tokens, n):
    # Contar la frecuencia de cada token en el corpus
    token_freq = Counter(tokens)

    # Filtrar los tokens que aparecen al menos 'N' veces
    filtered_tokens = [token for token in tokens if token_freq[token] >= n]

    return filtered_tokens

# Ejemplo de uso
corpus = drop_less_frecuency_words(corpus, 10)
print(corpus)
```

#### 7. **Dado el corpus, normalizado y con tokens que aparecen 10 veces o más en el corpus, se pide crear una bolsa de palabras en ONE-HOT-ENCODE con Gensim**

```python
!python3 -m nltk.downloader punkt
```

```python
import nltk
import gensim

tokenize = [nltk.word_tokenize(text) for text in corpus]
dictionary = gensim.corpora.Dictionary(tokenize)
vectors = [[(token[0], 1) for token in dictionary.doc2bow(doc)] for doc in tokenize]

# Resultados
print('Diccionario de palabras -> palabra: id\n')
print(dictionary.token2id)
print('\nApariciones de las palabras en los documentos (id, 1):')
vectors
```

#### 8. **Dado el corpus, normalizado y con tokens que aparecen 10 veces o más en el corpus, se pide crear una bolsa de palabras aplicando el TF-IDF con Scikit**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def create_tfidf_bag_of_words(corpus):
    # Convertir el corpus en una lista de documentos como strings
    corpus_texts = [' '.join(corpus)]

    # Inicializar el vectorizador TF-IDF
    tfidf_vectorizer = TfidfVectorizer()

    # Calcular TF-IDF para el corpus
    tfidf_matrix = tfidf_vectorizer.fit_transform(corpus_texts)

    # Obtener las palabras características (features)
    feature_names = tfidf_vectorizer.get_feature_names_out()

    return tfidf_matrix, feature_names

tfidf_matrix, feature_names = create_tfidf_bag_of_words(corpus)
# Mostrar el resultado
print("Matriz TF-IDF:")
print(tfidf_matrix.toarray())
print("\nPalabras características (features):")
print(feature_names)
```
